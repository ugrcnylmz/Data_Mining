# Veri Ön İşleme

```python
from sklearn.model_selection import train_test_split
# Veriyi eğitim ve test olarak ayırmak için kullanılır.
from sklearn.preprocessing import MinMaxScaler
# MinMax Scaling, verinin 0 ile 1 arasında değerler olduğu durumdur. 
```

```python
X = df.drop('Cancer', axis=1) # x te özellikleri tuttum.
y = df['Cancer'].values # y de hedef özellik mevcut.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, test_size = 0.10, random_state = 42)
```

```python
scaler = MinMaxScaler()
# Verileri oldukça dar bir aralığa topladığı ve daha hız kazandırdığı için MinMaxScaler kullanıldı.
```

```python
X_train= scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
# Veri leakage'i önlemek adına fit kısmını X_train ile, trasnform işlemini de X_test ile yaptım.
```

# Modelleme ve Model Performansı

```python
# Model oluşumu için gerekli kütüphaneler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation 
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import plot_roc_curve, roc_auc_score, roc_curve
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.model_selection import GridSearchCV
```

```python
X_train.shape
# train test olarak ayırdıktan sonra train de bulunan veri sayııs 512 oldu.
```

```python
X_test.shape
# train test olarak ayırdıktan sonra test te bulunan veri sayııs 57 oldu.
```

```python
model = Sequential()

model.add(Dense(30, activation = "relu"))  # diğerlerine göre öğrenme hızı daha iyi olduğu için relu kullanıldı.
model.add(Dense(15, activation = "relu"))
model.add(Dense(1, activation = "sigmoid"))  # output değerler binary olduğu için sigmoid kullandım. 

model.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"]) # adam optimizeri kullanıldı.
# Var olan 24 özellik için nöron sayısını 30 a yükselttim. İkinci katmanda 15 e düşürdüm. Çıktı da 1 olacak.
# Sigmoid aktivasyon fonksiyonu aynı zamanda logistic fonksiyondu ve veriyi 0 ile 1 arasına sıkıştırıyordu. Çıkış katmanı için bunu kullanmayı tercih ettim.
# Çıkış binary olduğundan, loss metriği olarak binarı_crossentropy kullandım.
```

```python
model.fit(x = X_train, y = y_train, validation_split = 0.1, batch_size = 32, epochs = 1000, verbose=1)

# Eğitim için kullanılacak datanın %10 luk bölümünü ayırdık. O veriyi eğitime sokmuyoruz. Validasyon işlemi için kullandım.
```

```python
model.summary()
# 1000 x 15 iterasyon oldu.
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.head()
# loss ve val_loss iterasyonlar ilerledikçe aşağıya doğru azaldı. 
# accuracy ve val_accurasy değeri de artıyor.
```

```python
loss_df.plot();

# 200 değerinden sonra overfit e gitmiş.
```

```python
model.evaluate(X_test, y_test, verbose=0)
```

```python
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print("loss : ", loss)
print("accuracy : ", accuracy)
```

```python
y_pred = (model.predict(X_test) > 0.5).astype("int32")   # 0.5 ten büyükse 1 yap
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# Earlystopping

```python
from tensorflow.keras.callbacks import EarlyStopping
# Earlystopping , aşırı öğrenmeyi önlemek için kullandığım bir yöntem.
```

```python
model = Sequential()

model.add(Dense(30, activation = "relu"))
model.add(Dense(15, activation = "relu"))
model.add(Dense(1, activation = "sigmoid"))

model.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
early_stop = EarlyStopping(monitor = "val_loss", mode = "auto", verbose = 1, patience = 15)

# EarlyStop kullanarak önceki aşırı öğrenmenin önüne geçeceğiz.
# Earlystop val_loss değerini monitör edecek.
# patience değerini 15 alarak daha iyi sonuçlar almayı hedefledim.
```

```python
model.fit(x = X_train, y = y_train, validation_split = 0.1, batch_size = 32, epochs = 1000, verbose = 1,
          callbacks = [early_stop]) 
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot();
# Earlystop kullanarak overfittinge girmeden eğitimi durdurmasını sağladık.
```

```python
y_pred = (model.predict(X_test) > 0.5).astype("int32")
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

# modelimin skorunu alırken train setindeki ayırdığım %10 ile değil, ilk datamdan ayırdığım test seti ile yapıyorum.
```

# Öğrenme Hızı (Learning Rate)

```python
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()

model.add(Dense(30, activation = "relu"))
model.add(Dense(15, activation = "relu"))
model.add(Dense(1, activation = "sigmoid"))

opt = Adam(lr = 0.005)
model.compile(optimizer = opt, loss = "binary_crossentropy", metrics = ["accuracy"])

# Öğrenme hızı veya adım büyüklüğü dediğimiz parametre büyük olursa minimum olan noktaya ulaşılamaz, çok küçük olursa da model çok yavaş öğrenir.
```

```python
early_stop = EarlyStopping(monitor = "val_loss", mode = "auto", verbose = 1, patience = 15)
```

```python
model.fit(x = X_train, y = y_train, validation_split = 0.1, batch_size = 32, epochs = 1000, verbose = 1,
          callbacks = [early_stop])
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot();
# Learning rate ile adımlarımı büyüttüğüm için beklediğim gibi skorda dalgalanma oldu.
```

```python
y_pred = (model.predict(X_test) > 0.5).astype("int32")
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# Dropout

```python
from tensorflow.keras.layers import Dropout
#  Eğitim sırasında aşırı öğrenmeyi engellemek için bazı nöronları unutmak için kullandım.
```

```python
model = Sequential()

model.add(Dense(30, activation="relu"))
model.add(Dropout(0.5))                   # Her seferinde rastgele şekilde nöronların yarısını kapatacak.
model.add(Dense(15, activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(1, activation="sigmoid"))

opt = Adam(lr = 0.001)
model.compile(optimizer = opt, loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
model.fit(x = X_train, y = y_train, validation_split = 0.1, batch_size = 32, epochs = 1000, verbose = 1)
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot()

# Grafiğe baktığımızda, earlystop kullanmadan sadece Dropout ile overfitting i düşürdük.
```

```python
y_pred = (model.predict(X_test) > 0.5).astype("int32")
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

 # Çapraz Geçerleme (Cross Validation)

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
```

```python
def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(units = 30, activation = 'relu'))
    classifier.add(Dense(units = 15, activation = 'relu'))
    classifier.add(Dense(units = 1, activation = 'sigmoid'))
    classifier.compile(optimizer = "adam", loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
```

```python
classifier = KerasClassifier(build_fn = build_classifier, batch_size = 32, epochs = 100, verbose=0)

scores = cross_validate(estimator = classifier, X = X_train, y = y_train,
                        scoring = ['accuracy', 'precision', 'recall', 'f1'], cv = 10)

df_scores = pd.DataFrame(scores, index = range(1, 11)).iloc[:, 2:]

df_scores_summary = pd.DataFrame({"score_mean" : df_scores.mean().values, "score_std" : df_scores.std().values},
                                 index = ["acc", "pre", "rec", "f1"])

print(df_scores_summary)
```

#GridSearchCV

```python
def build_classifier(optimizer):
    classifier = Sequential()
    classifier.add(Dense(units = 30, activation = 'relu'))
    classifier.add(Dense(units = 15, activation = 'relu'))
    classifier.add(Dense(units = 1, activation = 'sigmoid'))
    classifier.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
```

```python
early_stop = EarlyStopping(monitor = "accuracy", mode = "auto", verbose = 1, patience = 15)
```

```python
classifier = KerasClassifier(build_fn = build_classifier, epochs = 200)
parameters = {'batch_size': [32, 64],
              'optimizer': ['adam', 'rmsprop', "SGD", "adagrad", "adadelta"]}
grid_model = GridSearchCV(estimator = classifier,
                          param_grid = parameters,
                          scoring = 'accuracy',
                          cv = 10,
                          n_jobs = -1,
                          verbose = 1)
grid_model.fit(X_train, y_train, callbacks = [early_stop])
```

```python
grid_model.best_score_
```

```python
grid_model.best_params_
```
